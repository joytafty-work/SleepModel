```python
# Load models
from sklearn import datasets, linear_model
import pylab as pl
import numpy as np
diabetes = datasets.load_diabetes()

db_X = diabetes.data[:, np.newaxis]
db_X_temp = db_X[:, :, 2]
print db_X.shape 
print db_X_temp.shape

db_X_train = db_X_temp[:-20]
db_X_test = db_X_temp[-20:]
print db_X_train.shape
print db_X_test.shape

db_y_train = diabetes.target[:-20]
db_y_test = diabetes.target[-20:]
print db_y_train.shape
print db_y_test.shape

# Regression model 
regr = linear_model.Lasso(alpha = 0.1)
regr.fit(db_X_train, db_y_train)
print regr.coef_
print np.mean((regr.predict(db_X_test) - db_y_test)**2)
print regr.score(db_X_test, db_y_test)

pl.scatter(db_X_test, db_y_test, color='black')
pl.plot(db_X_test, regr.predict(db_X_test), color='red', linewidth=3)
pl.xticks()
pl.yticks()
plt.axes().yaxis.grid()
plt.axes().xaxis.grid()

pl.figure()
pl.scatter(db_y_test, regr.predict(db_X_test))
```
