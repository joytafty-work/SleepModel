```python
# Import dependencies
fpath = "static/data/UP/UPSleepHeader_noempty.csv"
import pandas as pd
import numpy as np
from itertools import izip
```

```python
# Read data
date = []
d = pd.read_csv(fpath)
del d['s_rem']
# change date format
for j in range(len(d)): 
    dd = str(d.date[j])
    out = str([dd[:4] + '-' + dd[5:7] + '-' + dd[-2:]])
    date.append([pd.to_datetime(out, "%Y-%m-%d-%H:%M")])
print d
```

```python
df = d.dropna()
df = df[df['s_quality'] > 0]
s = dict()
s['target'] = df['s_quality']
xcols = set(df.columns)
xcols.remove('s_quality')
xcols.remove('date')
# xcols.remove('s_rem')
s['data'] = df[list(xcols)]
phold = 0.25
nhold = int(np.ceil(phold*(s['target']).shape[0]))
```

```python
## Apply linear models
from sklearn import linear_model
from sklearn.linear_model import LassoCV
import pylab as pl
sX = array(s['data'])
sy = array(s['target'])
sy_train = sy[:-nhold]
sy_test = sy[-nhold:]
sX_train = sX[:-nhold]
sX_test = sX[-nhold:]
print (sX_train.shape, sy_train.shape, sX_test.shape, sy_test.shape)

# Lasso model simpler
clf = linear_model.Lasso(alpha=0.01)
clf.fit(sX_train, sy_train)
rsq = clf.score(sX_train, sy_train)
# print clf.sparse_coef_
ypred = clf.predict(sX_test)

pl.scatter(sy_test, ypred)
pl.xlabel('holdout sleep quality')
pl.ylabel('predicted sleep quality')
plt.axes().yaxis.grid()
plt.axes().xaxis.grid()

# LassoCV: coordinate descent
import time
t1 = time.time()
model = LassoCV(cv=20).fit(sX, sy)
t_lasso_cv = time.time() - t1

# Display results
# print model.alphas_
m_log_alphas = -np.log10(model.alphas_)
pl.figure()
pl.plot(m_log_alphas, model.mse_path_, '-')
plt.axes().yaxis.grid()
plt.axes().xaxis.grid()
pl.xlabel('-log(alpha)')
pl.ylabel('MSE')
pl.title('MSE each fold: coordinate descent, train time: %.2fs' % t_lasso_cv)
print model
```

```python
# Unsupervised clustering
import numpy as np
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(dat, quantile=0.2, n_samples=1000)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
datfit = dat_pca
ms.fit(datfit)
labels = ms.labels_
cluster_centers = ms.cluster_centers_
labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)

import pylab as pl
from itertools import cycle

pl.figure(1)
pl.clf()

colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
for k, col in zip(range(n_clusters_), colors):
    my_members = labels == k
    cluster_center = cluster_centers[k]
    pl.plot(datfit[my_members, 0], datfit[my_members, 1], col + 'o', markerfacecolor='w',
            markeredgecolor=col, linewidth=5)
    pl.plot(cluster_center[0], cluster_center[1], '*', markerfacecolor=col,
            markeredgecolor='k', markersize=20)
pl.title('Estimated number of clusters: %d' % n_clusters_)
pl.xlabel("PC1")
pl.ylabel("PC2")
plt.axes().yaxis.grid()
plt.axes().xaxis.grid()
pl.show()
```
